# 误差&复杂度

##### 数值微分误差

## 舍入误差

- 运算得到的近似值和精确值之间的差异，原因是数值类型有限位精度

```python
0.1+0.2
```

- 原因是：十进制0.1转成二进制是无限是无限循环小数

- 对python而言，浮点数精度大概为$10^{-16}$

在某些情况下，舍入误差会随着运算次数增加而累计很大，最终得到没有意义的运算结果

## 近似误差

- 数值计算公式是一个近似的公式

差分计算公式：

$${\frac{df}{dx}}=\lim_{h\to{0}}{\frac{f(x+h)-f(x)}{h}}$$

计算的步长h不可能无限小，导致误差

考虑一个典型的简谐运动方程：

$\ddot{x}=-x$

考虑一个正比于速度的阻尼，其方程有如下形式

$\ddot{x}=-x-\dot{x}$

可以证明，该方程有特解：

$x=e^{{\frac{-t}{2}}}\cos({\frac{\sqrt{3}}{2}}t)$

```python
import numpy as np
import matplotlib.pyplot as plt

# 位置随时间的函数
def xt(t):
    return np.exp(-t/2)*np.cos(np.sqrt(3)/2.*t)

# 速度随时间的函数（解析表达式）
def dxdt(t):
    return -0.5 * np.exp(-t/2) * np.cos(np.sqrt(3)/2.*t) \
        -0.5*np.sqrt(3)*np.exp(-t/2)* np.sin(np.sqrt(3)/2.*t)

# 数值计算差分
def numerical_diff(f, x, method=None, h=None):
    if method == 1: # 前差分
        return (f(x+h)-f(x)) / h
    elif method == 2: # 中心对称差分
        return (f(x+h)-f(x-h)) / (2*h)
    elif method == 3: # 五点差分
        return (-f(x+2*h)+8*f(x+h)-8*f(x-h)+f(x-2*h))/(12*h)
```

##### 数值微分与解析结果一致

```python
t = np.linspace(0, 20, 200)

plt.plot(t, dxdt(t), label='analytical')
plt.plot(t, numerical_diff(xt, t, method=1, h=1e-3), ls='--', lw=3, label='numerical')
plt.legend()
```

##### 三种差分，三种步长（1e-8, 1e-5, 1e-3，分别对应向前，中心，五点差分最佳步长）
    每种方法都在其最佳步长处取得最小的 Error
    五点 好于 中心 好于 向前

```python
t = np.linspace(0, 20, 200)

fig, axs = plt.subplots(3,3, figsize=(16,10))
method_labs = ['Forward', 'Central', 'Five-point']
for i_h, h in enumerate([1e-3, 1e-5, 1e-8]):
    for i_method, method in enumerate([1,2,3]):
        error = dxdt(t) - numerical_diff(xt, t, method=method, h=h)
        ax = axs[i_h, i_method]
        
        ax.plot(t, error, label=method_labs[i_method]+'\nh=%.0e'%h, lw=2)
        ax.legend(frameon=False, fontsize=14)
        if i_h == 2: ax.set_xlabel('t', fontsize=14)
        if i_method == 0: ax.set_ylabel('Error', fontsize=14)
```

# 时间复杂度

### 考量不同算法之间的优劣

#### 从算法所占用的时间和空间两个维度去考量！

- 时间维度：执行当前算法所消耗的时间，用“时间复杂度”来描述

- 空间维度：执行当前算法所需要占用多少内存空间，用“空间复杂度”来描述

##### 冒泡排序（时间复杂度O($N^{2}$)） 与 快速排序（时间复杂度O($N\log_{2}{N}$)）

```python
# 冒泡排序（Bubble Sort）是一种简单直观的排序算法。拿第一个和第二个进行相比，谁大就往后放。
# 它重复地走访过要排序的数列，一次比较两个元素，如果他们的顺序错误就把他们交换过来。

def bubble_sort(array):
    for i in range(len(array) - 1):
        swap = False #为假，没有拍好序
        for j in range(len(array) - 1 - i):
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]
                swap = True
        if not swap:
            break
    return array


#————————————————
#版权声明：本文为CSDN博主「欢喜躲在眉梢里」的原创文章，遵循CC 4.0 BY-SA版权协议，转载请附上原文出处链接及本声明。
#原文链接：https://blog.csdn.net/m0_52165864/article/details/125899863


# 快速排序的流程：
# (1) 首先设定一个分界值，以便把数组分为左、右两部分。
# (2) 将大于或等于分界值的数据集中在数组右边小于分界值的数据集中在数组的左边
# (3) 然后左边和右边的数据又可以独立排序取分界值将数据分为左右两部分不断重复上述过程
def partition(li,left,right):  
    tmp = li[left]
    while left < right:
        while left < right and li[right] >= tmp:  #从右边找比tmp小的数
            right -= 1   #继续从右往左查找
        li[left] = li[right]  #把右边的值写到左边空位上
        
        while left < right and li[left] <= tmp:
            left += 1
        li[right] = li[left] #把左边的值写到右边空位上
      
    li[left] = tmp #把tmp归位
    return  left

def quick_sort(li,left,right):  
    if left < right :#至少两个元素
        mid = partition(li,left,right)
        quick_sort(li,left,mid-1)
        quick_sort(li,mid+1,right)
#————————————————
        #版权声明：本文为CSDN博主「想成大神的大艳」的原创文章，遵循CC 4.0 BY-SA版权协议，转载请附上原文出处链接及本声明。
#原文链接：https://blog.csdn.net/qq_51201337/article/details/123322027
```

```python
np.random.seed(0)
lst = list(np.round(np.random.uniform(size=10, ),2))

print('Before sort:\n\t', lst)
print('\nBubble sort:\n\t', bubble_sort(lst))

np.random.seed(0)
lst = list(np.round(np.random.uniform(size=10, ),2))
quick_sort(lst, 0, 9)
print('\nQuick sort:\n\t', lst)
```

```python
import time

import sys
sys.setrecursionlimit(100000)   # 修改最大递归深度
                                # python的递归深度是有限制的，默认为1000。当递归深度超过1000时，就会报错.

size_lst = np.concatenate([np.arange(200, 1801, 200), np.arange(2000, 10001, 1000)])
bubble_time = []
quick_time = []

for size in size_lst:
    print('# size = ', size)
    
    np.random.seed(1)
    lst = list(np.random.uniform(size=size, ))
    
    t1 = time.perf_counter()
    bubble_sort(lst)
    t2 = time.perf_counter()
    bubble_time.append(t2-t1)
    print('\tBubble sort time cost: ', t2-t1)

    np.random.seed(1)
    lst = list(np.random.uniform(size=size, ))
    t1 = time.perf_counter()
    quick_sort(lst, 0, size-1)
    t2 = time.perf_counter()
    quick_time.append(t2-t1)
    print('\tQuick  sort time cost: ', t2-t1, '\n')
```

```python
bubble_time, quick_time = np.array(bubble_time), np.array(quick_time)

plt.plot(size_lst, bubble_time, marker='o', label='Bubble sort')
plt.plot(size_lst, quick_time,  marker='*', label='Quick sort ')
plt.xlabel('N (length of array)'); plt.ylabel('Time Cost')
plt.legend(fontsize=14)
plt.yscale('log')
plt.show();

plt.plot(size_lst, bubble_time / size_lst**2, marker='o', label='Bubble sort: t/N^2');
#plt.plot(size_lst, bubble_time / np.log(size_lst)/size_lst, marker='o', label='Bubble sort: t/(N log(N))');
#plt.plot(size_lst, quick_time / size_lst**2, marker='o', label='Quick sort: t/N^2')
plt.plot(size_lst, quick_time / np.log(size_lst)/size_lst, marker='o', label='Quick sort: t/ (N log(N))')
#plt.yscale('log')
plt.legend(fontsize=14);
```
